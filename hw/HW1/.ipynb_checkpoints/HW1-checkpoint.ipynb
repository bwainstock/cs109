#### Problem 1(d)

How would you graphically display the relationship between total wins and total salaries for a given year? What kind of plot would be best? Choose a plot to show this relationship and specifically annotate the Oakland baseball team on the on the plot. Show this plot across multiple years. In which years can you detect a competitive advantage from the Oakland baseball team of using data science? When did this end? 

**Hints**: Use a `for` loop to consider multiple years. Use the `teamID` (three letter representation of the team name) to save space on the plot.  

```python
years = wins_salaries_by_team.index.levels[1]
plt.style.use('ggplot')
fig, ax = plt.subplots(6,5, figsize=(25,20))
for year, axes in zip(years, ax.flatten()):
    sample = wins_salaries_by_team.loc[(wins_salaries_by_team.index.get_level_values('yearID')==year)]
    #Regression line
    (m, b) = scipy.polyfit(sample['salary'], sample['W'], 1)
    yp = scipy.polyval([m,b], sample['salary'])
    axes.plot(sample['salary'], yp)
    
    axes.scatter(y=sample['W'], x=sample['salary'])
    #axes.axhline(sample['W'].mean())
    #axes.axvline(sample['salary'].mean())
    #Oakland and New York labels
    xyoak = (sample['salary']['OAK'], sample['W']['OAK'])
    axes.annotate('OAK', xy=xyoak)
    xyny = (sample['salary']['NYA'], sample['W']['NYA'])
    axes.annotate('NYA', xy=xyny)
    #Settings
    axes.set_title(year)
    axes.set_xlabel('Salary')
    axes.set_ylabel('Wins')

ax[-1, -1].axis('off')
plt.tight_layout()
plt.show()
```

#### Problem 2(c)

Write a function to merge the `countries` and `income` data sets for any given year. 

```python
"""
Function
--------
mergeByYear

Return a merged DataFrame containing the income, 
country name and region for a given year. 

Parameters
----------
year : int
    The year of interest

Returns
-------
a DataFrame
   A pandas DataFrame with three columns titled 
   'Country', 'Region', and 'Income'. 

Example
-------
>>> mergeByYear(2010)
"""
def mergeByYear(year):
    income_to_merge = income[[year]]
    income_to_merge.columns = ['Income']
    joined = pd.merge(countries, income_to_merge, left_on=['Country'], right_index=True)
    joined['Income'] = np.round(joined['Income'], 2)
    return joined
    
mergeByYear(2010)
```

#### Problem 2(d) 

Use exploratory data analysis tools such as histograms and boxplots to explore the distribution of the income per person by region data set from 2(c) for a given year. Describe how these change through the recent years?

**Hint**: Use a `for` loop to consider multiple years.  

```python
years = range(1950, 2010, 10)
fig, ax = plt.subplots(1, len(years), figsize=(25,5))
for year, axes in zip(years, ax):
    income_by_year = mergeByYear(year)
    income_by_year.boxplot('Income', by='Region', rot=90, figsize=(5,5), ax=axes)
    #plt.ylim(10**2, 10.5 **5)
    axes.set_yscale('log')
    axes.set_ylim(top=10**5)
    axes.set_title(year)
    axes.set_ylabel('Income per person (log10 scale)')
plt.tight_layout()
```

```python
a = mergeByYear(1950).groupby('Region').mean()
b = mergeByYear(2000).groupby('Region').mean()

print('Change in mean income:', b-a)
```

#### Problem 3(a)

Assume you have two list of numbers, X and Y, with distribution approximately normal. X and Y have standard deviation equal to 1, but the average of X is different from the average of Y.  If the difference in the average of X and the average of Y is larger than 0, how does the proportion of X > a compare to the proportion of Y > a? 

Write a function that analytically calculates the ratio of these two proportions: Pr(X > a)/Pr(Y > a) as function of the difference in the average of X and the average of Y. 

**Hint**: Use the `scipy.stats` module for useful functions related to a normal random variable such as the probability density function, cumulative distribution function and survival function. 

**Update**: Assume Y is normally distributed with mean equal to 0. 

Show the curve for different values of a (a = 2,3,4 and 5).

```python
"""
Function
--------
ratioNormals

Return ratio of these two proportions: 
    Pr(X > a)/Pr(Y > a) as function of 
    the difference in the average of X 
    and the average of Y. 

Parameters
----------
diff : difference in the average of X 
    and the average of Y. 
a : cutoff value

Returns
-------
Returns ratio of these two proportions: 
    Pr(X > a)/Pr(Y > a)
    
Example
-------
>>> ratioNormals(diff = 1, a = 2)
"""
import scipy.stats as stats

def ratioNormals(diff, a):
    X = stats.norm(loc=diff)
    Y = stats.norm()
    prob_x = 1 - X.cdf(a)
    prob_y = 1 - Y.cdf(a)
    return prob_x/prob_y
```

```python
a = range(2,6)
diffs = np.linspace(0,5,50)

for val in a:
    ratios = [ratioNormals(diff, val) for diff in diffs]
    plt.plot(diffs, ratios)

#Legends
plt.title('Pr(X>a)/Pr(Y>a)')
plt.xlabel('a')
plt.ylabel('Probability ratio')
plt.yscale('log')
plt.legend(["a={}".format(val) for val in a], loc=0);
```
